```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
C = 5*10**-6  # C = 5 microF
R = 2*10**3       # R = 2 kilo-ohm
ur = 0         
tr = 5*10**-3   #tr =5 ms
v = 1*10**-3   # v = 1mV
TC = R*C*1000        # TC = 10ms
```

```python
#1.  I = 2uA for 10ms <= t <= 75ms; I = 0Amp otherwise
h = 0.1   # Time step = h = 0.1ms
I = np.zeros(int(100/h))
I[int(10/h):int(75/h)] = 2*10**-6        # I = 2microA
u = np.zeros(int(100/h))
u_grad = np.zeros(int(100/h))

    
```

```python
t = 101
while t>int(0) and t<int(100/h-1):
    u_grad[t] = -(u[t]/TC) + (I[t]/C)        #Euler
    u[t+1] = u[t] + h*u_grad[t]
    t = t + 1
    if u[t]>1:
        count = 1
        u[t:t+49] = 0
        t = t+49
```

```python
t = np.arange(100, step = 0.1)
plt.plot(t, u)
plt.title('membrane potential (u) vs. time graph')
plt.xlabel('Time Unit 1ms ')
plt.ylabel('Voltage 1mV ')
plt.show()
```
